```python


# -------------------------
# PRICE PARSING UTILITIES
# -------------------------

def indian_to_number(x):
    if pd.isna(x):
        return np.nan

    x = str(x).replace("₹", "").replace(",", "").strip().lower()

    # handle ranges
    if "-" in x:
        parts = [p.strip() for p in x.split("-")]
        nums = [indian_to_number(p) for p in parts]
        return min(nums), max(nums)

    # unit conversion
    if x.endswith("k") or x.endswith("k."):
        return float(x.replace("k", "").replace("k.", "")) * 1e3

    if x.endswith("l") or x.endswith("lac") or x.endswith("lakh"):
        return float(re.sub("[^0-9.]", "", x)) * 1e5

    if x.endswith("cr") or x.endswith("crore"):
        return float(re.sub("[^0-9.]", "", x)) * 1e7

    # plain digits
    try:
        return float(re.sub("[^0-9.]", "", x))
    except:
        return np.nan


# ------------------------------------
# NORMALIZE ALL PRICE COLUMNS
# ------------------------------------

price_cols = [c for c in df.columns if "bhk flat" in c.lower() or "sq.yd plot" in c.lower() \
    or "sq.ft plot" in c.lower() or "studio apartment" in c.lower()]

for col in price_cols:
    df[f"{col}_min"] = np.nan
    df[f"{col}_max"] = np.nan

    for i, val in df[col].items():
        parsed = indian_to_number(val)

        if isinstance(parsed, tuple):
            df.at[i, f"{col}_min"] = parsed[0]
            df.at[i, f"{col}_max"] = parsed[1]
        else:
            df.at[i, f"{col}_min"] = parsed
            df.at[i, f"{col}_max"] = parsed


df.drop(price_cols, axis=1, inplace=True)



# -------------------------
# AREA NORMALIZATION
# -------------------------

def parse_area(col_name, value):
    if pd.isna(value):
        return np.nan

    # extract numeric
    try:
        area = float(re.sub("[^0-9.]", "", value))
    except:
        return np.nan
    
    # detect unit by column name
    name = col_name.lower()

    if "yd" in name or "yard" in name:
        return area * 0.836127
    elif "ft" in name or "sq.ft" in name:
        return area * 0.092903
    elif "m" in name and "sq" in name:
        return area
    else:
        return np.nan


area_cols = [c for c in df.columns if "builtup_area" in c.lower()]

for col in area_cols:
    df[f"{col}_m2"] = df[col].apply(lambda x: parse_area(col, x))


# ------------------------------------
# OPTIONAL: MELT THE AREA INTO LONG FORM
# ------------------------------------

area_long = df.melt(
    id_vars=['project'],
    value_vars=[c for c in df.columns if c.endswith("_m2")],
    var_name="area_type",
    value_name="area_m2"
)

# Similarly for prices (if needed)
```

```python
# import pandas as pd
# import numpy as np
# import re

# # load your df
# df = pd.read_csv('/mnt/data/2_df.csv')

def _parse_single_value_in_inr(text: str):
    """Parse a single piece like '2.89k', '4.5 k', '777.0' to INR (float)."""
    t = text.strip().lower()
    # detect 'k' (thousands)
    if 'k' in t:
        num = re.sub(r'[^0-9.]', '', t)
        try:
            return float(num) * 1_000
        except ValueError:
            return np.nan
    else:
        num = re.sub(r'[^0-9.]', '', t)
        try:
            return float(num)
        except ValueError:
            return np.nan

def parse_avg_price_to_inr_per_sqm(x):
    """Convert strings like '₹4.5 K/sq.ft' or '₹3.02 K - ₹4.5 K/sq.ft'
       into integer INR per sq.meter."""
    if pd.isna(x):
        return np.nan
    
    s = str(x).strip().lower()

    # Filter out non-price strings like 'dec, 2025'
    if not any(ch.isdigit() for ch in s):
        return np.nan

    # Extract area unit: ft / yd / m / meter / mtr / mt
    unit_match = re.search(r'/\s*sq\.?\s*(ft|yd|m|meter|mtr|mt)', s)
    if unit_match:
        unit = unit_match.group(1)
    else:
        # no recognizable unit => skip
        return np.nan

    # Clean currency symbol & commas, remove the "/sq.xxx" part
    s_clean = s.replace('₹', '').replace(',', '')
    s_clean = re.sub(r'/\s*sq\.?\s*(ft|yd|m|meter|mtr|mt)', '', s_clean).strip()

    # Handle ranges: "a - b"
    if '-' in s_clean:
        parts = [p.strip() for p in s_clean.split('-') if p.strip()]
        nums = []
        for p in parts:
            nums.append(_parse_single_value_in_inr(p))
        nums = [n for n in nums if n is not None and not np.isnan(n)]
        if not nums:
            return np.nan
        value_per_unit = sum(nums) / len(nums)  # average of range
    else:
        value_per_unit = _parse_single_value_in_inr(s_clean)

    if value_per_unit is None or np.isnan(value_per_unit):
        return np.nan

    # Convert per-unit → per-square-meter
    if unit == 'ft':
        value_per_sqm = value_per_unit / 0.092903
    elif unit == 'yd':
        value_per_sqm = value_per_unit / 0.836127
    else:  # m / meter / mtr / mt -> already per m²
        value_per_sqm = value_per_unit

    # Return rounded integer INR / sq.m
    return int(round(value_per_sqm))


# Apply to your dataframe
df['avg_price_inr_per_sqm'] = df['avg_price'].apply(parse_avg_price_to_inr_per_sqm)

# Quick check
print(df[['avg_price', 'avg_price_inr_per_sqm']].head(10))
```
